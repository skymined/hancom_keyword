## 데이터 전처리

```python
# 영어, 한글만 선택
import re

def replace_text(text):
    text = text.replace('\\n',' ')
    text = re.sub(r'[^a-zA-Z가-힣\s]',' ',text)
    return text   
```

```python
df_news['newText'] = df_news['news_title'] + ' ' + df_news['news_contents']
```

```python
df_news['newText'] = df_news['newText'].apply(replace_text)
```

```python
news_brand = list(df_news['news_brand'].unique())
```

```python
# 뉴스 브랜드, 기자 단어 제거

temp_list = []

for i in df_news['newText']:
    for j in news_brand:
        i = i.replace(j,'')
        i = i.replace('기자','')
    temp_list.append(i)

df_news['newText'] = temp_list
```

```python
import pickle

with open('./pickles/date_dataframe.pkl', "rb") as file:
    df_date = pickle.load(file)
```

```python
df_date = df_date.sort_values()
```

```python
# 분석해야될 날짜의 뉴스 갯수 카운트

from collections import Counter

count_dict = Counter(df_news['news_date'])
```

```python
from pandas import Timestamp

date_replace_dic = {}

for k, v in count_dict.items():
    # Timestamp 객체 생성
    timestamp_obj = Timestamp(year=k.year, month=k.month, day=k.day)
    # Timestamp 객체를 저장
    date_replace_dic[timestamp_obj] = v

date_replace_dic
```

```python
cal_count_dic={}

for i in df_date:
    cal_count_dic[i] = date_replace_dic.get(i,0)

cal_count_dic
```

```python
df_check_date = pd.DataFrame(cal_count_dic.items(), columns=['date','count'])
df_check_date
```

- 뉴스는 2018 ~ 2023년까지 수집하였으므로 그 외 기간 날짜는 drop

```python
df_check_date = df_check_date.query('date < "2024-01-01"')
```

```python
import seaborn as sns

sns.lineplot(df_check_date, x='date',y='count')
```
