```python
import pandas as pd
```

```python
news_read1 = pd.read_csv('../../selenium/yohan/crawling/news_get.daum_news.csv')
news_read2 = pd.read_csv('../../selenium/sky/crawling/local.daum_news.csv')
news_read3 = pd.read_csv('../../selenium/juhyeon/crawling/daum_news_2020-2021.csv')

df = [news_read1,news_read2,news_read3]
```

```python
df_news = pd.concat(df)
df_news
```

```python
# 여러번에 걸쳐 수집이 되었으므로 중복값을 드랍
df_news = df_news.drop(columns='_id')
df_news = df_news.drop_duplicates()
df_news
```

```python
# 뉴스 브랜드 확인
df_news['news_brand'].unique()
```

- 이상치로 '매일경', '머니', 'Copyrigh' 항목이 존재
- '매일경' = '매일경제', '머니' = '머니투데이', 'Copyrigh' = '전자신문' 으로 변경

```python
df_news.loc[df_news['news_brand'] == '매일경', 'news_brand'] = '매일경제'
df_news.loc[df_news['news_brand'] == '머니', 'news_brand'] = '머니투데이'
df_news.loc[df_news['news_brand'] == 'Copyrigh', 'news_brand'] = '전자신문'
```

```python
df_news.info()
```

- date값이 object타입인것을 확인 -> date타입으로 변경
- 크롤링한 주식의 단위가 일일 단위이므로 시간은 삭제

```python
df_news['news_date'] = pd.to_datetime(df_news['news_date']).dt.date
```
